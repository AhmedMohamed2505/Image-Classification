```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import array_to_img, img_to_array, load_img
```

```python
data_generator = ImageDataGenerator(rotation_range=50,
                                   zoom_range=0.3,
                                   shear_range=0.3,
                                   horizontal_flip=True,
                                   brightness_range=(0.5, 2))
```

```python
image = load_img('/kaggle/input/xxcxczxczxcxzcxzcxzc/OIP.png')
x = img_to_array(image)
x = x.reshape((1, ) + x.shape)
```

```python
import os
os.makedirs('/kaggle/working/preview', exist_ok=True)
```

```python
i = 0
for batch in data_generator.flow(x, 
                                 batch_size=2,
                                 save_to_dir='preview',
                                 save_prefix='image',
                                 save_format='jpeg'):
    i += 1
    if i >= 5:   
        break
```

```python
import matplotlib.pyplot as plt
```

```python
for i in os.listdir('preview'):
  plt.imshow(plt.imread(os.path.join('preview',i)))
  plt.show()
```
